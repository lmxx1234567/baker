```python
from difflib import SequenceMatcher

def similar(string: str, sub: str) -> float:
    sm = SequenceMatcher(None, string, sub)
    matchs = sm.get_matching_blocks()
    size = 0
    for match in matchs:
        size+=match.size
    return size/len(sub)
```

```python
similar('2、关于隋文江按何种标准对待问题。','隋文江何种标准对待')
```
